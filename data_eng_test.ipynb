```python
path = "US_youtube_trending_data.csv"
path_category = "US_category_id.json"
```

```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv(path)
df_category = pd.read_json(path_category)
```

```python
df
```

```python
df_sumView = df.groupby("categoryId").sum()
df_sumView
```

```python
df_sorted = df_sumView.sort_values("view_count", ascending=False)
df_sorted = df_sorted.head(10)
df_sorted
```

```python
df_category
```

```python
column_json = ['id', 'title', 'kind', 'etag']
```

```python
df_name = pd.DataFrame(columns = column_json)
df_name
```

```python
ls_items = []
for index,json in df_category.iterrows():
  ls_items.append(json['items'])
  df_name = df_name.append({'id' : json['items']['id'], 'kind' : json['items']['kind'], 'etag' : json['items']['etag'], 'title' : json['items']['snippet']['title']}, ignore_index=True)
df_title = df_name
df_title
```

```python
result = df_sorted.merge(df_title, left_index=True, right_index=True)
final_result = result.sort_values("view_count", ascending=False)
final_result
```

```python
final_result.drop(columns=['likes','dislikes','comment_count','comments_disabled','ratings_disabled','id','kind','etag','index'], inplace=True)
final_result
```
